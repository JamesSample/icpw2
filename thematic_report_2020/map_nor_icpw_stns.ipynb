```python
%matplotlib inline

import nivapy3 as nivapy
import pandas as pd
import matplotlib.pyplot as plt
```

# Map Norwegian ICPW stations

```python
xl_path = r'../data/all_icpw_sites_may_2019.xlsx'
df = pd.read_excel(xl_path, sheet_name='all_icpw_stns')
df = df.query('country == "Norway"')
df.head()
```

```python
nivapy.spatial.quickmap(df,
                        popup='station_code',
                        aerial_imagery=True,
                        kartverket=True,
                        )
```
